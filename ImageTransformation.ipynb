```python
import numpy as np
import matplotlib.pyplot as plt


heart_img = np.array([[255,0,0,255,0,0,255],
              [0,255/2,255/2,0,255/2,255/2,0],
          [0,255/2,255/2,255/2,255/2,255/2,0],
          [0,255/2,255/2,255/2,255/2,255/2,0],
              [255,0,255/2,255/2,255/2,0,255],
                  [255,255,0,255/2,0,255,255],
                  [255,255,255,0,255,255,255]])

# This is a helper function that makes it easy for you to show images!
def show_image(image, name_identifier, cmap="gray"):
  plt.imshow(image, cmap=cmap)
  plt.title(name_identifier)
  plt.show()

# Show heart image

show_image(heart_img, "Heart Image")
# Invert color
inverted_heart_img = 255 - heart_img

show_image(inverted_heart_img, "Inverted Heart")



# Rotate heart
rotated_heart_img = heart_img.T
show_image(rotated_heart_img, "Rotated Heart Image")

# Random Image
random_img = np.random.randint(0,255,(7,7))

show_image(random_img, "Random Image")

# Solve for heart image

X = np.linalg.solve(random_img, heart_img)

show_image(X, "X")

solved_heart_img = random_img@X

show_image(solved_heart_img, "Solved Heart Image.")


# Create a new shape
new_shape_img = np.array([
    [0, 0, 255, 0, 0],
    [0, 255, 255, 255, 0],
    [255, 255, 255, 255, 255],
    [0, 255, 255, 255, 0],
    [0, 0, 255, 0, 0]
])
show_image(new_shape_img, "New Shape")

# Apply a permutation matrix to transform the heart image
P = np.roll(np.eye(7), shift=1, axis=1)  # Shift columns cyclically
permuted_heart = P @ heart_img
show_image(permuted_heart, "Permuted Heart Image")

# Change the color scheme
show_image(heart_img, "Heart in Plasma", cmap="plasma")

# Create a more complex, larger image
large_heart_img = np.random.randint(0, 255, (10, 15))
show_image(large_heart_img, "Larger Random Image")

# Create a non-square complex image
complex_image = np.random.randint(0, 255, (20, 30))
show_image(complex_image, "Complicated Non-Square Image")

# Create a simplified pixel art version of Goku
goku_pixel_art = np.array([
    [0, 0, 255, 255, 255, 0, 0],
    [0, 255, 255, 255, 255, 255, 0],
    [255, 255, 0, 255, 0, 255, 255],
    [255, 255, 255, 255, 255, 255, 255],
    [0, 255, 255, 255, 255, 255, 0],
    [0, 0, 255, 255, 255, 0, 0],
    [0, 0, 0, 255, 0, 0, 0]
])
show_image(goku_pixel_art, "Simplified Goku Pixel Art")
```
